```python
import numpy as np
import time
import matplotlib.pyplot as plt

NUM_SAMPLES = 500

np.random.seed(int(time.time()))              # 난수의 생성패턴을 정함.
#np.random.seed(0) 이면 항상 동일한 패턴이 발생하게 됨.

xs = np.random.uniform(-2, 0.5, NUM_SAMPLES)  # pick randome. in the range.
print("xs = ", xs)
np.random.shuffle(xs)                         # in random order
print("xs in random order = ", xs)

ys = 2*xs**2 + 3*xs + 5
print("ys = ", ys[:5])

plt.plot(xs, ys, 'b.')
plt.show()
'''
z = np.random.randn(NUM_SAMPLES)
print("z = ", z)
plt.hist(z)
plt.show()
'''
ys += 0.2 * np.random.randn(NUM_SAMPLES)
plt.plot(xs, ys, 'g.')  # green
plt.show()

NUM_SPLIT = int(0.8 * NUM_SAMPLES)

x_train, x_test = np.split(xs, [NUM_SPLIT])
y_train, y_test = np.split(ys, [NUM_SPLIT])

plt.plot(x_train, y_train, 'b.', label = 'train') # blue
plt.plot(x_test, y_test, 'r.', label = 'test')    # red
plt.legend()
plt.show()

import tensorflow as tf

model_f = tf.keras.Sequential([
                               tf.keras.layers.InputLayer(input_shape=(1,)),
                               tf.keras.layers.Dense(16, activation='relu'),
                               tf.keras.layers.Dense(16, activation='relu'),
                               tf.keras.layers.Dense(1)
])

model_f.compile(optimizer='rmsprop', loss='mse')

p_test = model_f.predict(x_test)

plt.plot(x_test, y_test, 'b.', label='actual')
plt.plot(x_test, p_test, 'r.', label='predicted')
plt.legend()
plt.show()

#학습
model_f.fit(x_train, y_train, epochs=60)

p_test = model_f.predict(x_test)

plt.plot(x_test, y_test, 'b.', label='actual')
plt.plot(x_test, p_test, 'r.', label='predicted')
plt.legend()
plt.show()
```

```python
import tensorflow as tf

mnist = tf.keras.datasets.fashion_mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(" x_train:%s \n y_train:%s \n x_test:%s \n y_test:%s \n"
%(x_train.shape, y_train.shape, x_test.shape, y_test.shape))

import matplotlib.pyplot as plt
plt.figure()
plt.imshow(x_train[0])
plt.show()


for y in range(28):
  for x in range(28):
    print("%4s" %x_train[0][y][x], end=' ')  # 자동 \n 막음
  print("\n")

class_names = [
               'T-shirt/top',
               'Trouser',
               'Pullover',
               'Dress',
               'Coat',
               'Sandal',
               'Shirt',
               'Sneaker', 
               'Bag',
               'Ankle boot'
]

plt.figure(figsize=(10,10))
for i in range(25):
  plt.subplot(6,7,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(x_train[i], cmap=plt.cm.binary)
  plt.xlabel(class_names[y_train[i]], fontsize=15)
plt.show()


x_train, x_test = x_train / 255.0, x_test / 255.0
x_train, x_test = x_train.reshape((60000, 784)), x_test.reshape((10000, 784))

model = tf.keras.Sequential([
                             tf.keras.layers.InputLayer(input_shape=(784,)),
                             tf.keras.layers.Dense(128, activation='relu'),
                             tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])

model.fit(x_train, y_train, epochs = 5)

model.evaluate(x_test, y_test)

p_test = model.predict(x_test)

import numpy as np
print('p_test[0]: ', np.argmax(p_test[0]), class_names[np.argmax(p_test[0])], 'y_test[0]: ', y_test[0], class_names[y_test[0]])

x_test = x_test.reshape(10000, 28, 28)

plt.figure()
plt.imshow(x_test[0])
plt.show()


#plt.figure(figsize=(50,50))
plt.figure(figsize=(50,100))
for i in range(100):
  #plt.subplot(20,7, i+1)  # total length is already fixed. 50/20=2.5
  plt.subplot(20,7, i+1)  # total length is already fixed. 50/20=2.5
  plt.xticks([])
  plt.yticks([])
  plt.imshow(x_test[i], cmap=plt.cm.binary)
  plt.xlabel("%s / %s" %(np.argmax(p_test[i]), class_names[np.argmax(p_test[i])]), fontsize=15)
plt.show()


cnt_wrong = 0
p_wrong = []
for i in range(10000):
  if np.argmax(p_test[i]) != y_test[i]:
      p_wrong.append(i)
      cnt_wrong = cnt_wrong + 1

cnt_wrong = min(100, cnt_wrong) #100개만 표시

column_count = 10
plt.figure(figsize=(30, 6*(cnt_wrong/5)))  # 전체세로길이를 개당인치높이*개수로.
for k in range(cnt_wrong):
  plt.subplot(cnt_wrong/5 + 1, 5, k+1) # row, column   # what if all wrong.
  plt.xticks([])
  plt.yticks([])
  print("count=", cnt_wrong, "k=", k, "p_wrong[k]=", p_wrong[k])
  plt.imshow(x_test[p_wrong[k]], cmap=plt.cm.binary)
  plt.xlabel("p=%s / y=%s" 
             %(
                 class_names[np.argmax(p_test[p_wrong[k]])],
                 class_names[y_test[p_wrong[k]]]
             ), 
             fontsize=15)
plt.show()
# Chpa_01 Ends.
```
